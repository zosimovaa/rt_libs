```python
# Tests for dynamic plot updates
#-------------------------------
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import time

x = np.linspace(0, 10*np.pi, 100)
y = np.sin(x)

# The really important command for interactive plot updating
plt.ion()

# sizing of the plots figure sizes 
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 8
fig_size[1] = 3

# Two figures 
# -----------
fig1 = plt.figure(1)
fig2 = plt.figure(2)

# first figure with two plot-areas with axes 
# --------------------------------------------
ax1_1 = fig1.add_subplot(121)
ax1_2 = fig1.add_subplot(122)

fig1.canvas.draw()

# second figure with just one plot area with axes
# -------------------------------------------------
ax2 = fig2.add_subplot(121)
line1, = ax2.plot(x, y, 'b-')
fig2.canvas.draw()

z= 32
b = np.zeros([1])
c = np.zeros([1])
c[0] = 1000

for i in range(z):
    # update data 
    phase = np.pi / z * i 
    line1.set_ydata(np.sin(0.5 * x + phase))
    b = np.append(b, [i**2])
    c = np.append(c, [1000.0 - i**2])
    
    # re-plot area 1 of fig1  
    ax1_1.clear()
    ax1_1.set_xlim (0, 100)
    ax1_1.set_ylim (0, 1000)
    ax1_1.plot(b)
    
    # re-plot area 2 of fig1  
    ax1_2.clear()
    ax1_2.set_xlim (0, 100)
    ax1_2.set_ylim (0, 1000)
    ax1_2.plot(c)
    
    # redraw fig 1 
    fig1.canvas.draw()

    # redraw fig 2 with updated data  
    fig2.canvas.draw()
    
    time.sleep(0.1)
```

```python
class TrainRender:
    FIG_SIZE_X = 14
    FIG_SIZE_Y = 5
    DPI=50
    
    def __init__(self, config):
        self.config = config
        self.data = dict()
        self.lines = dict()
        plt.ion()
        
        self.n_plots = len(self.config)
        self.fig, self.ax = plt.subplots(figsize=(self.FIG_SIZE_X, self.FIG_SIZE_Y*self.n_plots), 
                                         nrows=self.n_plots, 
                                         dpi=self.DPI, 
                                         constrained_layout=True)
        #self.fig, self.ax = plt.subplots(nrows=n_plots)
        if self.n_plots ==1:
            self.ax = [self.ax]

        for i in range(self.n_plots):
            if "primary" in self.config[i]:
                for key in self.config[i]["primary"].keys():
                    self.config[i]["primary"][key], = self.ax[i].plot([], [])

            if "secondary" in self.config[i]:
                for key in self.config[i]["secondary"].keys():
                    self.config[i]["secondary"][key], = self.ax[i].twinx().plot([], [])

        self.fig.canvas.draw()
        
    def __save_sample(self, sample):
        for metric in sample.keys():
            if metric not in self.data:
                self.data[metric] = []
            self.data[metric].append(sample[metric])
    
    def update_plot(self, sample):
            
        self.__save_sample(sample)

        for i in range(len(self.config)):
            if "primary" in self.config[i]:
                for key in self.config[i]["primary"].keys():
                    self.config[i]["primary"][key].set_ydata(self.data[key])
                    self.config[i]["primary"][key].set_xdata(range(len(self.data[key])))
                    
                    self.config[i]["primary"][key].axes.relim() 
                    self.config[i]["primary"][key].axes.autoscale_view(True,True,True) #Autoscale
            
            if "secondary" in self.config[i]:
                for key in self.config[i]["secondary"].keys():
                    self.config[i]["secondary"][key].set_ydata(self.data[key])
                    self.config[i]["secondary"][key].set_xdata(range(len(self.data[key])))
                    
                    self.config[i]["secondary"][key].axes.relim() 
                    self.config[i]["secondary"][key].axes.autoscale_view(True,True,True) #Autoscale

        self.fig.canvas.draw()
        self.fig.canvas.flush_events()
        
        
        
        
        
            
```
